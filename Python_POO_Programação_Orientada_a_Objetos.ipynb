## Aula 05 - Melhorando Classes e Criando uma Conta Bancária

```python
# Declaração de Classe
class Gafanhoto2:
  # Docstring
  """
  Essa classe cria um Gafanhoto, que é uma pessoa que tem nome e idade.

  Para criar uma nova pessoa, use:
  variavel = Gafanhoto2(nome, idade)
  """
  def __init__(self, nome = "vazio", idade = 0): # Método Construtor
    # Atributos de Instância
    self.nome = nome
    self.idade = idade
  # Método de Instância
  def aniversario(self):
    self.idade = self.idade + 1

  def mensagem(self):
    return f"{self.nome} é Gafanhoto(a) e tem {self.idade} anos."

  def __str__(self): # Dunder Method
    return f"{self.nome} é Gafanhoto(a) e tem {self.idade} anos."

  def __getstate__(self):
    return f"Estado: nome = {self.nome} ; idade = {self.idade}"

# Declaração de Objetos
g1 = Gafanhoto2("Maria", 17)
g1.aniversario()
#print(g1.mensagem())
print(g1)

g2 = Gafanhoto2("Mauro", 53)
g2.aniversario()
#print(g2.mensagem())
print(g2)

g3 = Gafanhoto2()
#print(g1.mensagem())
print(g3)


print(g1.__doc__) # Dunder Attribute
print(g1.__dict__) # Attribute
print(g1.__getstate__()) # Method
print(g2.__getstate__()) # Method
print(g1.__class__) # Dunder Attribute
```

```python
from rich import print
from rich import inspect

# Declaração de Classe
class ContaBancaria:
  """
  Cria uma conta bancária e permite fazer saques e depósitos
  """
  def __init__(self, id, nome, saldo = 0): # Mátodo Construtor
    # Atritutos de Instância
    self.id = id
    self.titular = nome
    self.saldo = saldo
    print(f"Conta {self.id} criada com sucesso. Saldo R${self.saldo:,.2f}")
  # Método de Instância
  def __str__(self):
    return f"A conta {self.id} de {self.titular} tem R${self.saldo:,.2f} de saldo."

  def depositar(self, valor):
    self.saldo += valor
    print(f"Depoósito de R${valor:.2f} autorizado na conta {self.id}.")

  def sacar(self, valor):
    if valor > self.saldo:
      print(f"Saldo insuficiente! Saldo em conta R${self.saldo:,.2f}")
    else:
      self.saldo -= valor
      print(f"Saque de R${valor:,.2f} autorizado na conta {self.id}.")

# Declaração de Objetos
c1 = ContaBancaria(122, "Gustavo", 3000)
print(c1)
#c1.depositar(500)
#print(c1)
#c1.sacar(20000)
#print(c1)
#print(c1.__doc__)

c2 = ContaBancaria(111, "Jośe", 500)
inspect(c2)
```

```python
"""
- Desafio 17 - V1
Crie uma classe Produto, onde podemos cadastrar: nome e o preço.
Crie também um método que mostre uma etiqueta do preço do produto.
"""
from rich import print
from rich.panel import Panel

# Declaração de Classe
class Produto:
  def __init__(self, nome, preco = 0): # Método construtor
    # Atributos de Instância
    self.nome = nome
    self.preco = preco
  # Método de Instância
  def etiqueta(self):
    precof = f"R${self.preco:,.2f}"
    conteudo = f"{self.nome:^25}\n{26*'-'}\n{precof:.^25}"
    etiqueta = Panel(conteudo, width=30, title="Produto")
    print(etiqueta)

# Declaração de Objetos
p1 = Produto("iphone 17 Pro Max", 25000)
p2 = Produto("Notebook Gamer", 8000)
p1.etiqueta()
p2.etiqueta()
```

```python
"""
- Desafio 17 - V2
Crie uma classe Produto, onde podemos cadastrar: nome e o preço.
Crie também um método que mostre uma etiqueta do preço do produto.
"""
from rich import print
from rich.panel import Panel

# Declaraçao de Classe
class Produto:
  def __init__(self, nome, preco = 0): # Método construtor
    # Atributos de Instância
    self.nome = nome
    self.preco = preco
  # Método de Instância
  def etiqueta(self):
    conteudo = f"{self.nome.center(25, ' ')}"
    conteudo += f"{'-' * 26}"
    precof = f"R${self.preco:,.2f}"
    conteudo += f"{precof.center(26, '.')}"
    etiqueta = Panel(conteudo, width=30, title="Produto")
    print(etiqueta)

# Declaração de Objetos
p1 = Produto("iphone 17 Pro Max", 25000)
p2 = Produto("Notebook Gamer", 8000)
p1.etiqueta()
p2.etiqueta()
```

```python
"""
- Desafio 18 - V1
Crie uma classe Churrasco, onde seja possível informar quantas pessoas
vão participar e mostre quanto de carne deve ser comprado, o custo total do
churrasco e o preço por pessoa.
CONSIDERE:
Consumo padrão: 400g por pessoa
Preço: R$ 82,40/kg
"""
from rich import print
from rich.panel import Panel
# Declração de classe
class Churrasco:
  def __init__(self, titulo, qtd): # Método construtor
    # Atributos de Instância
    self.titulo = titulo
    self.qtd = qtd

  # Método de Instância
  def analisar(self):
    custo_kg_carne = 82.40
    kg_p_pessoa = 0.40
    custo_total = (custo_kg_carne * kg_p_pessoa) * self.qtd
    recom = custo_total / custo_kg_carne
    intro = f"Analisando [green]{self.titulo}[/] com [dodger_blue2]{self.qtd} convidados[/].\n"
    intro += f"Cada participante comerá 0.4kg e cada kg custa R$ 82.40.\n"
    intro += f"Recomendo [dodger_blue2]comprar {recom:,.2f}Kg[/] de carne.\n"
    intro += f"O custo total será de [green]R${custo_total:,.2f}[/].\n"
    intro += f"Cada pessoa pagará [yellow]R${custo_kg_carne * kg_p_pessoa}[/] para participar."
    texto = Panel(intro, width=70, title=self.titulo)
    print(texto)

# Declaração de Objetos
c1= Churrasco("Churras dos Amigos", 15)
c1.analisar()
```

```python
"""
- Desafio 19
Crie uma classe Livro, que vai simular a passgem de páginas de um livro,
considerando também se o usuário chegou ao fim da leitura.
"""
from rich import print
from rich.console import Console
# Delcaração de Classes
class Livro:
  def __init__(self, titulo, paginas): # Método Construdor
    # Atributos de Instância
    self.titulo = titulo
    self.total_paginas = paginas
    self.pagina_atual = 1

    print(f"""
  :book: Você acabou de abrir o livro '{self.titulo}' que tem {self.total_paginas} páginas
  no total. Você agora está na [yellow]página {self.pagina_atual}[/].
  Você avançou páginas e agora está na página {self.pagina_atual}.
      """) # remove espaços extras no começo/fim

  # Método de Instância
  def avancar_paginas(self, qtd):
    self.contador = 0
    while self.contador < qtd:
      if self.pagina_atual < self.total_paginas:
        self.pagina_atual += 1
        self.contador += 1
        print(f"Pág.{self.contador}")
    self.pagina_atual = self.contador
    print(f"Você avançou {self.contador} páginas e agora está na página {self.pagina_atual}.")


# Declaração de objetos
l1 = Livro("10 coisas que aprendi", 85)
l1.avancar_paginas(5)
l1.avancar_paginas(10)



```
